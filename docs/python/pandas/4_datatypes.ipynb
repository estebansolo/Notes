## Estructuras de dataframes en detalle

```python
df_meteorites.nunique()
```

```python
# Varaibles Categoricas (Ya que no se repiten sino que definen alguna categoria)
# Trabajar con variables categoricas reduce el uso de memoria ram y tamaÃ±o de archivos
df_meteorites[["fall", "nametype"]]
```

```python
# Define una nueva variable en el dataframe
df_meteorites[["fall", "nametype"]] = df_meteorites[["fall", "nametype"]].astype("category")
```

```python
df_meteorites.dtypes
```

```python
df_meteorites["fall"].unique()
```

```python
# Frecuencia de estas categorias
df_meteorites['fall'].value_counts()
```

```python
# Pasa de ser formato texto a ser formato numerico
pd.get_dummies(df_meteorites["fall"])
```

```python
df_meteorites["ones"] = 1
df_meteorites
```

```python
df_meteorites[["fell", "found"]] = pd.get_dummies(df_meteorites["fall"])
df_meteorites
```

```python
df_meteorites["year"]
```

```python
pd.to_datetime(df_meteorites["year"], errors="coerce", format="%m/%d/%Y %H:%M:%S %p")
```
