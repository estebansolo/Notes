# Algoritmos de Ordenamiento

Son algoritmos que se encargan de darle orden a un set de datos o elementos (Numeros o Letras).

## Ejemplo

- `[a, b, c, d]`: Un arreglo ordenado alfabeticamente.
- `[1, 2, 3, 4]`: Arreglo de numeros enteros ordenados ascendentemente.

## Bubble Sort

El algoritmo de ordenamiento mas simple, Se llama asi porque selecciona los elementos adyacentes y los pone en orden correcto (los va ordenando de 2 en 2).

- Es muy facil
- Es muy lento
- Lo hace de 2 en 2
- Por lo general se usa con 2 ciclos

### BigO Notation

`O(n^2)`



```python
%%time

# 1. Comparar elementos adyacentes.
# 2. Repetir hasta tener un ciclo sin cambios.

def bubble_sort(array):
    n = len(array)
    
    for i in range(n):
        print(array)
        has_changes = False
        
        for j in range(n - i - 1):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
                has_changes = True
        
        if not has_changes:
            break
            
    return array

unorganized = [4603,37,76,5259,4356,9,1010,5382,20,427]
organized = bubble_sort(unorganized)

print ("\nEl arreglo ordenado de forma Ascendente es: ", end="")

print(", ".join([str(n) for n in organized]))
    
```

## Selection Sort

Su funcionamiento es el siguiente:

- Buscar el mínimo elemento de la lista
- Intercambiarlo con el primero
- Buscar el siguiente mínimo en el resto de la lista
- Intercambiarlo con el segundo

Y en general:

- Buscar el mínimo elemento entre una posición i y el final de la lista
- Intercambiar el mínimo con el elemento de la posición i

### BigO Notation

`O(n^2)`




```python
%%time

# 1. Buscar el numero menor en el array
# 2. Crear 2 sub arrays para llevar el control del algoritmo

import sys 

def selection_sort(array):
    for i in range(len(array)): 
        # Find the minimum element in remaining unsorted array 
        min_idx = i
        
        for j in range(i + 1, len(array)): 
            if array[min_idx] > array[j]: 
                min_idx = j 

        # Swap the found minimum element with the first element         
        array[i], array[min_idx] = array[min_idx], array[i] 
    
    return array

unorganized = [4603,37,76,5259,4356,9,1010,5382,20,427]
organized = selection_sort(unorganized)

print ("\nEl arreglo ordenado de forma Ascendente es: ", end="")

print(", ".join([str(n) for n in organized]))
```
